```python

# ---------------------------
# 0) Install / import packages
# ---------------------------

# If running in Colab uncomment the following installs:
# !pip install pandas numpy scikit-learn matplotlib seaborn tensorflow imbalanced-learn

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import (precision_score, recall_score, f1_score, accuracy_score,
                             roc_auc_score, confusion_matrix, classification_report)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ---------------------------
# 1) Load dataset (SWaT)
# ---------------------------

# NOTE: Replace the path below with your SWaT csv file path after uploading to Colab or mounting Drive.
# Common SWaT filenames: "SWaT_Dataset_Attack_v0.csv", "SWaT_Dataset_Normal_v0.csv" or combined versions.

DATA_PATH = '/content/drive/MyDrive/PROJECT DATASET/SWaT_Dataset_Attack_v0.xlsx'  # update this path

if not os.path.exists(DATA_PATH):
    print("WARNING: dataset not found at", DATA_PATH)
    print("Please upload the SWaT dataset into the Colab session and update DATA_PATH")
else:
    print("Found dataset at", DATA_PATH)

# Attempt to read file robustly
try:
    df = pd.read_excel(DATA_PATH, header=1)
    print('Loaded dataset shape:', df.shape)
    display(df.head())
except Exception as e:
    print('Error loading dataset:', e)
    df = None

# ---------------------------
# 2) Quick EDA & preprocessing
# ---------------------------

# The SWaT dataset typically contains many sensor columns and an "Attack" or "Label" column.
# We'll try to auto-detect a label column. If not present, you'll need to supply one.

if df is not None:
    # lower-case columns
    df.columns = [c.strip() for c in df.columns]
    # try to find label column
    label_cols = [c for c in df.columns if c.lower() in ['label','attack','is_attack','attack_type','state']]
    print('Possible label columns:', label_cols)

    if len(label_cols) > 0:
        label_col = label_cols[0]
        print('Using label column:', label_col)
    else:
        # fallback: sometimes SWaT has 'Normal/Attack' in a file split; ask user to provide
        print('No obvious label column found. Please ensure dataset contains a label column and update script.')
        label_col = None

    # Basic cleaning: drop non-numeric time cols (if any 'Timestamp') and parse timestamp if exists
    time_cols = [c for c in df.columns if 'time' in c.lower() or 'date' in c.lower() or 'timestamp' in c.lower() or 't_stamp' in c.lower()]
    print('Time-like columns:', time_cols)
    if len(time_cols) > 0:
        time_col = time_cols[0]
        try:
            # Address UserWarning: Parsing dates with dayfirst=True for better handling of dd/mm/yyyy format
            df[time_col] = pd.to_datetime(df[time_col], dayfirst=True)
            df = df.sort_values(time_col).reset_index(drop=True)
        except Exception:
            pass
    # Select numeric sensor columns as features
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print('Numeric columns detected (sample):', numeric_cols[:10])
    if label_col and label_col in numeric_cols:
        # label is numeric, keep it separate
        numeric_cols = [c for c in numeric_cols if c != label_col]

    # If label column exists but not numeric, map values
    if label_col and label_col in df.columns and df[label_col].dtype == object:
        df[label_col] = df[label_col].apply(lambda x: 1 if str(x).lower().strip()!='normal' and str(x).lower().strip()!='0' else 0)

    # Final feature matrix X and labels y (if available)
    if len(numeric_cols) == 0:
        raise ValueError('No numeric feature columns found. Please preprocess dataset to include sensors as numeric columns.')

    X_all = df[numeric_cols].copy()
    y_all = df[label_col].copy() if label_col and label_col in df.columns else None

    print('Feature matrix shape:', X_all.shape)

    # Simple missing value handling
    missing_summary = X_all.isnull().mean().sort_values(ascending=False)
    print('Top missing features fraction:\n', missing_summary.head())
    # Fill or drop columns with too many missing values
    thresh = 0.5
    cols_to_drop = missing_summary[missing_summary > thresh].index.tolist()
    if len(cols_to_drop) > 0:
        print('Dropping high-missing columns:', cols_to_drop)
        X_all = X_all.drop(columns=cols_to_drop)

    # For remaining missing values, forward-fill then mean
    # Address FutureWarning: DataFrame.fillna with 'method' is deprecated
    X_all = X_all.ffill().fillna(X_all.mean())

    # Scale features
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X_all), columns=X_all.columns)

else:
    raise FileNotFoundError('Dataset not loaded. Upload SWaT dataset and rerun.')

# ---------------------------
# 3) Train / validation / test split
# ---------------------------

# For time-series ICS data use time-based split if timestamp exists
if 'time_col' in locals() and time_col in df.columns:
    # use first 60% normal data for training (assumes normal period available), next 20% val, last 20% test
    n = len(df)
    train_end = int(0.6 * n)
    val_end = int(0.8 * n)
    X_train = X_scaled.iloc[:train_end]
    X_val = X_scaled.iloc[train_end:val_end]
    X_test = X_scaled.iloc[val_end:]
    if y_all is not None:
        y_train = y_all.iloc[:train_end]
        y_val = y_all.iloc[train_end:val_end]
        y_test = y_all.iloc[val_end:]
else:
    # random split fallback
    if y_all is not None:
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_all, test_size=0.2, random_state=42, stratify=y_all)
        X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42, stratify=y_train)
    else:
        print('No labels found, performing simple random split without stratification.')
        X_train_full, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
        X_train, X_val = train_test_split(X_train_full, test_size=0.25, random_state=42)
        y_train, y_val, y_test = None, None, None # Ensure labels are None if not available

print('Train/Val/Test shapes:', X_train.shape, X_val.shape, X_test.shape)

# Important: for unsupervised training of IsolationForest & LSTM autoencoder, prefer using normal-only samples in training
if y_all is not None:
    # Use only normal samples for unsupervised training
    normal_label = 0
    X_train_unsup = X_train[y_train == normal_label]
    print('Unsupervised training using only normal samples shape:', X_train_unsup.shape)
else:
    # if no labels, use whole training split
    X_train_unsup = X_train

# ---------------------------
# 4) Model A: Isolation Forest
# ---------------------------

iso = IsolationForest(n_estimators=200, contamination=0.02, random_state=42)
iso.fit(X_train_unsup)

# anomaly score: negative_outlier_factor-like -> map to 0/1
scores_iso_test = -iso.decision_function(X_test)  # higher = more anomalous
# derive binary preds using threshold: use percentile from val set
scores_iso_val = -iso.decision_function(X_val)
thresh_iso = np.percentile(scores_iso_val, 98)  # top 2% anomalous
y_pred_iso = (scores_iso_test >= thresh_iso).astype(int)

print('IsolationForest threshold (val 98th pct):', thresh_iso)

# ---------------------------
# 5) Model B: LSTM Autoencoder (sequence model)
# ---------------------------

# LSTM autoencoder expects sequences. We'll build sequences of length seq_len
seq_len = 10
features = X_train_unsup.shape[1]

# utility to create sequences
def create_sequences(X, seq_len=10):
    sequences = []
    for start in range(0, len(X) - seq_len + 1):
        seq = X.iloc[start:start+seq_len].values
        sequences.append(seq)
    return np.array(sequences)

# prepare normal sequences for training
seq_train = create_sequences(X_train_unsup, seq_len=seq_len)
print('LSTM train sequences shape:', seq_train.shape)

# validation and test sequences (we need aligned labels per sequence: if any sample in seq is anomalous -> label 1)
def seq_labels_from_point_labels(y_series, seq_len=10):
    y = y_series.reset_index(drop=True)
    seq_labels = []
    for start in range(0, len(y) - seq_len + 1):
        lbl = 1 if y[start:start+seq_len].sum() > 0 else 0
        seq_labels.append(lbl)
    return np.array(seq_labels)

seq_val = create_sequences(X_val, seq_len=seq_len)
seq_test = create_sequences(X_test, seq_len=seq_len)

if 'y_val' in locals() and y_val is not None:
    y_seq_val = seq_labels_from_point_labels(y_val, seq_len=seq_len)
    y_seq_test = seq_labels_from_point_labels(y_test, seq_len=seq_len)
else:
    y_seq_val = None
    y_seq_test = None

# Build LSTM autoencoder
model = models.Sequential([
    layers.Input(shape=(seq_len, features)),
    layers.LSTM(128, return_sequences=True),
    layers.Dropout(0.2),
    layers.LSTM(64, return_sequences=False),
    layers.RepeatVector(seq_len),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.2),
    layers.LSTM(128, return_sequences=True),
    layers.TimeDistributed(layers.Dense(features))
])

model.compile(optimizer='adam', loss='mse')
model.summary()

es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(seq_train, seq_train, epochs=50, batch_size=64, validation_data=(seq_val, seq_val) if len(seq_val)>0 else None, callbacks=[es])

# compute reconstruction error on test sequences
reconstructions = model.predict(seq_test)
mse = np.mean(np.power(seq_test - reconstructions, 2), axis=(1,2))

# determine threshold on val mse
reconstructions_val = model.predict(seq_val)
mse_val = np.mean(np.power(seq_val - reconstructions_val, 2), axis=(1,2))
thresh_lstm = np.percentile(mse_val, 98)
print('LSTM autoencoder threshold:', thresh_lstm)

y_pred_lstm_seq = (mse >= thresh_lstm).astype(int)

# Map sequence predictions back to point-level predictions (simple approach: mark center point of sequence)
# We'll build point-level anomaly scores by assigning sequence score to the seq center

def seq_to_point_scores(scores_seq, n_points, seq_len=10):
    point_scores = np.zeros(n_points)
    counts = np.zeros(n_points)
    for i, s in enumerate(scores_seq):
        for j in range(i, i+seq_len):
            point_scores[j] += s
            counts[j] += 1
    counts[counts==0] = 1
    return point_scores / counts

scores_lstm_test_point = seq_to_point_scores(mse, len(X_test), seq_len=seq_len)
# produce point-level binary predictions with threshold
# use same scalar threshold
# For consistency convert point scores threshold similarly (any aggregated score >= thresh -> anomaly)
y_pred_lstm_point = (scores_lstm_test_point >= thresh_lstm).astype(int)

# ---------------------------
# 6) Hybrid Model: stacking using anomaly scores
# ---------------------------

# Build features: isolation score + lstm sequence MSE score per point
# Compute scores for val set similarly
scores_iso_val_point = scores_iso_val  # note: iso decision function computed pointwise already

# For lstm we need val point scores
mse_val_seq = mse_val
scores_lstm_val_point = seq_to_point_scores(mse_val_seq, len(X_val), seq_len=seq_len)

# Ensure arrays align in length: val/test lengths may differ due to sequence windows.
min_len = min(len(scores_iso_val_point), len(scores_lstm_val_point), len(X_val))
X_stack_val = np.vstack([scores_iso_val_point[:min_len], scores_lstm_val_point[:min_len]]).T
if y_val is not None:
    y_stack_val = y_val.iloc[:min_len].values
else:
    y_stack_val = None

# Fit logistic regression on validation anomaly scores to predict labels
stack_clf = LogisticRegression()
if y_stack_val is not None:
    stack_clf.fit(X_stack_val, y_stack_val)
    print('Stacking model coefficients:', stack_clf.coef_, stack_clf.intercept_)
else:
    print('No labels available for stacking; hybrid will use average score ensemble instead.')

# Prepare test stacking features (align sizes)
min_len_test = min(len(scores_iso_test), len(scores_lstm_test_point), len(X_test))
X_stack_test = np.vstack([scores_iso_test[:min_len_test], scores_lstm_test_point[:min_len_test]]).T
if y_test is not None:
    y_stack_test = y_test.iloc[:min_len_test].values
else:
    y_stack_test = None

if y_stack_val is not None:
    y_pred_stack = stack_clf.predict(X_stack_test)
else:
    # average score ensemble -> threshold by percentile of average val score
    avg_val_score = np.mean(X_stack_val, axis=1)
    avg_test_score = np.mean(X_stack_test, axis=1)
    thresh_avg = np.percentile(avg_val_score, 98)
    y_pred_stack = (avg_test_score >= thresh_avg).astype(int)

# ---------------------------
# 7) Evaluation (compare models)
# ---------------------------

def evaluate(y_true, y_pred, name='Model'):
    print('\n=== Evaluation for', name, '===')
    if y_true is None:
        print('No ground truth labels available; cannot compute metrics.')
        return None
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    try:
        auc = roc_auc_score(y_true, y_pred)
    except Exception:
        auc = None
    print('Accuracy: %.4f | Precision: %.4f | Recall: %.4f | F1: %.4f | ROC-AUC: %s' % (acc, prec, rec, f1, ('%.4f' % auc) if auc else 'N/A'))
    print('Confusion Matrix:\n', confusion_matrix(y_true, y_pred))
    return dict(accuracy=acc, precision=prec, recall=rec, f1=f1, auc=auc)

# align y_test for each model
# For iso, we have y_test (point-level) already
if 'y_test' in locals() and y_test is not None:
    eval_iso = evaluate(y_test.iloc[:len(y_pred_iso)].values, y_pred_iso, name='IsolationForest')
    eval_lstm = evaluate(y_test.iloc[:len(y_pred_lstm_point)].values, y_pred_lstm_point, name='LSTM Autoencoder')
    eval_stack = evaluate(y_stack_test, y_pred_stack, name='Hybrid (Stacking)')
else:
    print('No ground truth labels to evaluate. Provide labeled SWaT dataset for metrics.')

# ---------------------------
# 8) Plots
# ---------------------------

plt.figure(figsize=(12,4))
plt.plot(scores_iso_test[:500], label='Iso score (first 500)')
plt.plot(scores_lstm_test_point[:500], label='LSTM point mse (first 500)')
plt.legend()
plt.title('Anomaly Scores Comparison (sample)')
plt.show()

# Confusion heatmap for hybrid
if y_stack_test is not None:
    cm = confusion_matrix(y_stack_test, y_pred_stack)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title('Confusion matrix - Hybrid')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# ---------------------------
# 9) Save models and produce outputs requested by assignment
# ---------------------------

# Save trained models (isolation forest and LSTM weights)
import joblib
joblib.dump(iso, 'isolation_forest_model.joblib')
model.save('lstm_autoencoder_model.h5')

# Save a CSV summary of evaluation metrics if computed
metrics = []
if 'eval_iso' in locals() and eval_iso:
    eval_iso['model']='IsolationForest'
    metrics.append(eval_iso)
if 'eval_lstm' in locals() and eval_lstm:
    eval_lstm['model']='LSTM_AE'
    metrics.append(eval_lstm)
if 'eval_stack' in locals() and eval_stack:
    eval_stack['model']='Hybrid'
    metrics.append(eval_stack)

if len(metrics)>0:
    pd.DataFrame(metrics).to_csv('evaluation_metrics_summary.csv', index=False)
    print('Saved evaluation_metrics_summary.csv')

print('All done. Export this notebook as .ipynb for submission and also use the report template provided below to generate Overleaf PDF.')

# ---------------------------
```
